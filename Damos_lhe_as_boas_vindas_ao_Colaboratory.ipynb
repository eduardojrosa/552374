```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
# Carregar dados
df = pd.read_csv('health_data.csv')
# Selecionar variáveis relevantes
df = df[['Cardiovascular Disease', 'Hypertension', 'Smoking', 'High Cholesterol', 'Sedentary']]
# Separar dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(df.drop('Cardiovascular Disease', axis=1), df['Cardiovascular Disease'], test_size=0.3, random_state=42)
# Treinar modelo de classificação
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
# Fazer previsões
train_pred = clf.predict(X_train)
test_pred = clf.predict(X_test)
# Avaliar modelo
train_acc = accuracy_score(y_train, train_pred)
test_acc = accuracy_score(y_test, test_pred)
print('Acurácia no treino:', train_acc)
print('Acurácia no teste:', test_acc)
```
